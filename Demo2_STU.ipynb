## Reading Data Subset


```python
# Import Geopandas and Pandas
import geopandas as gpd
import pandas as pd
```

```python
# Read the Countries data (similar to a previous demo)
shp = 'data/countries.shp'
countries = gpd.read_file(shp)
countries.head()
```

```python
# Plotting
countries.plot()
```

## Reading subset of data using Geometry

```python
# Can you filter out the data from the above frame?
# For example how can we retrieve the data corresponding to only Europe?
#Let's filter using a continent
europe = countries[countries["CONTINENT"] == "Europe"]
```

```python
# How can you plot only the filtered data?
# Plot the filtered data

europe.plot()
```

```python
# Can you filter the ConflictData.shp data by considering only Middle East?
# Reading Conflict Dataset by subseting Middle East
conflict_subset = gpd.read_file('data/ConflictData.shp')
```

```python
middle_east = conflict_subset[conflict_subset["region"] == "Middle East"]
middle_east.head()
```

```python
middle_east.plot()
```

## Reading subset of data using Rows

```python
# Slice the data through inputting a range of rows
Conflict_subset = gpd.read_file(
    "data/ConflictData.shp",

    rows=slice(100, 200)
)
Conflict_subset.plot();
```
